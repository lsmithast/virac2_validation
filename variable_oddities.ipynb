# Variable Oddities
Here I'll store interesting lightcurves I encounter while validating the catalogue.

```python
%matplotlib inline
import matplotlib.pyplot as plt
# import some general virac v2 lightcurve tools
from wsdb_lc import *

# we'll be doing a lot of plotting lightcurves, may as well make it a function
def plot_lc(sourceid, xlims=(2010,2019)):
    lc = LightCurve(sourceid)
    lc.to_years()
    plt.figure(figsize=(12,3))
    plt.errorbar(lc.t, lc.mag, yerr=lc.emag,
                 fmt='.', markersize=3, linewidth=1, capsize=2)
    plt.xlabel("t  /  %s" % lc.t_unit)
    plt.ylabel("K$_s$  /  mag")
    plt.xlim(*xlims)
    plt.gca().invert_yaxis()
    plt.grid()
    plt.tight_layout()
```
