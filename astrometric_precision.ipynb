# Astrometric Accuracy

```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import numpy as np
import wsdb
```

```python
# select approx. 1/100th of the sources that have parallax measurements at random
sql = "select mag, epm, eplx from virac_pm2 where rand_flt<0.01 and eplx>0"
data = wsdb.getsql(sql)
```

## Proper Motion Uncertainties

```python
plt.figure(figsize=(12,10))
plt.hist2d(data['mag'], np.log10(data['epm']), bins=400, cmap='YlOrRd', norm=LogNorm())
plt.xlabel('Ks Magnitude', fontsize=17)
plt.ylabel('Proper Motion Error ($mas\ yr^{-1}$)', fontsize=17)
yt = [0.1,0.2,0.5,1,2,5,10,20,50]
plt.ylim(np.log10(0.1), np.log10(100))
plt.yticks(np.log10(yt), ["%.1f"%y for y in yt])
plt.grid()
plt.tight_layout()
plt.savefig("mag_epm.eps", bbox_inches='tight')
```

## Parallax Uncertainties

```python
plt.figure(figsize=(12,10))
plt.hist2d(data['mag'], np.log10(data['eplx']), bins=400, cmap='YlOrRd', norm=LogNorm())
plt.xlabel('Ks Magnitude', fontsize=17)
plt.ylabel('Parallax Error ($mas$)', fontsize=17)
yt = [0.2,0.5,1,2,5,10,20,50,100]
plt.ylim(np.log10(0.2), np.log10(120))
plt.yticks(np.log10(yt), ["%.1f"%y for y in yt])
plt.grid()
plt.tight_layout()
plt.savefig("mag_eplx.eps", bbox_inches='tight')
```
